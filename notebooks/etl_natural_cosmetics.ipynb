```python
import pandas as pd
```

```python
customers = pd.read_csv("tables/customers_natural_cosmetics.csv")
customers.head()
```

```python
products = pd.read_csv("tables/products_natural_cosmetics.csv")
products.head()
```

```python
sales = pd.read_csv ("tables/sales_natural_cosmetics.csv")
sales ['sale_date'] = pd.to_datetime (sales['sale_date'], errors = 'coerce')
print (sales.info())
```

```python
sales = sales.sort_values('sale_date').reset_index(drop=True)
sales['sale_id'] = range(1, len(sales) + 1)
sales.tail(20)
```

```python
customers.info()
print (customers.isnull().sum())
```

```python
products.info()
print (products.isnull().sum())
```

```python
sales.info()
print (sales.isnull().sum())
```

```python
full_data = sales.merge(products, on = "product_id", how = "left")\
.merge(customers, on = "customer_id", how = "left") 
full_data.head()
```

```python
full_data = full_data.drop(['product_id', 'customer_id'], axis=1)
```

```python
full_data ['total_revenue'] = round(full_data['price'] * full_data['quantity'] * (1 - full_data['discount_percent']), 2)
full_data ['gross_revenue'] = round(full_data['price'] * full_data['quantity'], 2)
full_data ['discount_value'] = round(full_data['price'] * full_data['quantity'] * full_data['discount_percent'], 2)
full_data.head()
```

```python
full_data.to_csv("full_data_clean.csv", index=False)
```
